```python
from sklearn.model_selection import train_test_split

X = df[['Food Product','Main Ingredient', 'Sweetener', 'Fat/Oil', 'Seasoning']]
y = df['Prediction']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the training data
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train['Food Product'] + ' ' + X_train['Main Ingredient'] + ' ' + X_train['Sweetener'] + ' ' + X_train['Fat/Oil'] + ' ' + X_train['Seasoning'])

# Transform the test data
X_test_tfidf = tfidf_vectorizer.transform(X_test['Food Product'] + ' ' + X_test['Main Ingredient'] + ' ' + X_test['Sweetener'] + ' ' + X_test['Fat/Oil'] + ' ' + X_test['Seasoning'])

# Initialize the SVM classifier
svm_classifier = SVC(kernel='linear')

# Train the SVM classifier
svm_classifier.fit(X_train_tfidf, y_train)

# Predict the labels for the test set
y_pred = svm_classifier.predict(X_test_tfidf)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split



# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the training data
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train['Food Product'] + ' ' + X_train['Main Ingredient'] + ' ' + X_train['Sweetener'] + ' ' + X_train['Fat/Oil'] + ' ' + X_train['Seasoning'])

# Transform the test data
X_test_tfidf = tfidf_vectorizer.transform(X_test['Food Product'] + ' ' + X_test['Main Ingredient'] + ' ' + X_test['Sweetener'] + ' ' + X_test['Fat/Oil'] + ' ' + X_test['Seasoning'])

# Initialize the KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)  # You can adjust the number of neighbors as needed

# Train the KNN classifier
knn_classifier.fit(X_train_tfidf, y_train)

# Predict the labels for the test set
y_pred = knn_classifier.predict(X_test_tfidf)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB

tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the training data
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train['Food Product'] + ' ' + X_train['Main Ingredient'] + ' ' + X_train['Sweetener'] + ' ' + X_train['Fat/Oil'] + ' ' + X_train['Seasoning'])

# Transform the test data
X_test_tfidf = tfidf_vectorizer.transform(X_test['Food Product'] + ' ' + X_test['Main Ingredient'] + ' ' + X_test['Sweetener'] + ' ' + X_test['Fat/Oil'] + ' ' + X_test['Seasoning'])

# Initialize the Gaussian Naive Bayes classifier
gnb_classifier = GaussianNB()

# Train the Gaussian Naive Bayes classifier
gnb_classifier.fit(X_train_tfidf.toarray(), y_train)

# Predict the labels for the test set
y_pred = gnb_classifier.predict(X_test_tfidf.toarray())

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
